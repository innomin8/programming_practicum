```python
import pygame
import random

pygame.init()

screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
screen.fill((255, 255, 255))
pygame.display.set_caption('Первая программа в Pygame')

pygame.draw.circle(screen, 'red', [200, 100], 30, width=0)
pygame.draw.circle(screen, [255, 154, 13], [100, 400], 50, width=15)
pygame.draw.circle(screen, '#FFEE54', [400, 300], 100, width=5)

pygame.draw.rect(screen, 'yellow', [400, 20, 300, 200], 0)

for _ in range(5):
    top = random.randint(50, 700)
    left = random.randint(50, 500)
    width = random.randint(10, 200)
    height = random.randint(10, 100)
    color = [random.randint(0, 255) for _ in range(3)]
    pygame.draw.rect(screen, color, [top, left, width, height], 4)

dots = [
    [221, 432], [225, 331], [133, 342], [141, 310],
    [51, 230], [74, 217], [58, 153], [114, 164],
    [123, 135], [176, 190], [159, 77], [193, 93],
    [230, 28], [267, 93], [301, 77], [284, 190],
    [327, 135], [336, 164], [402, 153], [386, 217],
    [409, 230], [319, 310], [327, 342], [233, 331],
    [237, 432]
]
pygame.draw.lines(screen, 'green', True, dots, 2)

pygame.time.delay(2000)
pygame.draw.rect(screen, 'white', [400, 450, 100, 100], 0)
pygame.display.flip()

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
pygame.init()

screen = pygame.display.set_mode((800, 600))
screen.fill((255, 255, 255))
pygame.display.set_caption('Домик из линий')

center_x = 400
center_y = 300

house_width = 200
house_height = 150
roof_height = 100

house_left = center_x - house_width // 2
house_top = center_y - house_height // 2
house_rect = [house_left, house_top, house_width, house_height]

roof_top = (center_x, center_y - house_height // 2 - roof_height)
roof_left = (house_left, house_top)
roof_right = (house_left + house_width, house_top)
roof_points = [roof_top, roof_left, roof_right]

pygame.draw.rect(screen, 'brown', house_rect, 0)

pygame.draw.polygon(screen, 'red', roof_points, 0)

pygame.display.flip()

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption('Анимация фигур')

WHITE = (255, 255, 255)
COLORS = [
    (255, 0, 0),
    (0, 255, 0),
    (0, 0, 255),
    (255, 255, 0),
    (0, 255, 255),
    (255, 0, 255),
]

class Shape:
    def __init__(self, x, y, width, height, color, shape_type, speed):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.color = color
        self.shape_type = shape_type
        self.speed = speed

    def draw(self):
        if self.shape_type == "square":
            pygame.draw.rect(screen, self.color, (self.x, self.y, self.width, self.height))
        elif self.shape_type == "rectangle":
            pygame.draw.rect(screen, self.color, (self.x, self.y, self.width, self.height))
        elif self.shape_type == "circle":
            pygame.draw.circle(screen, self.color, (self.x + self.width // 2, self.y + self.height // 2), self.width // 2)
        elif self.shape_type == "triangle":
            points = [
                (self.x + self.width // 2, self.y),
                (self.x, self.y + self.height),
                (self.x + self.width, self.y + self.height)
            ]
            pygame.draw.polygon(screen, self.color, points)

    def move(self):
        self.x += self.speed
        if self.x <= 0 or self.x + self.width >= WIDTH:
            self.speed = -self.speed
            self.color = random.choice(COLORS)

    def is_clicked(self, mouse_pos):
        if self.shape_type == "circle":
            distance = ((mouse_pos[0] - (self.x + self.width // 2)) ** 2 + 
                        (mouse_pos[1] - (self.y + self.height // 2)) ** 2) ** 0.5
            return distance <= self.width // 2
        else:
            return (self.x <= mouse_pos[0] <= self.x + self.width and
                    self.y <= mouse_pos[1] <= self.y + self.height)

shapes = [
    Shape(100, 200, 50, 50, random.choice(COLORS), "square", 5),
    Shape(300, 300, 80, 40, random.choice(COLORS), "rectangle", 4),
    Shape(500, 100, 50, 50, random.choice(COLORS), "circle", 3),
    Shape(700, 400, 60, 60, random.choice(COLORS), "triangle", 6)
]

running = True
clock = pygame.time.Clock()
while running:
    screen.fill(WHITE)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            mouse_pos = pygame.mouse.get_pos()
            for shape in shapes:
                if shape.is_clicked(mouse_pos):
                    shape.color = random.choice(COLORS)

    for shape in shapes:
        shape.move()
        shape.draw()

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
